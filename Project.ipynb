### Data Science Project about Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population) - World

```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Reading data about Global Poverty
df = pd.read_csv('API_SI.POV.DDAY_DS2_en_csv_v2_3359022.csv')
df.head()
```

```python
#Transforming columns to rows
df = df.transpose()
```

```python
#Viewing the 5 first rows in a dataset
df.head()
```

```python
#Using the first row to dataset header
df.columns = df.iloc[0]
```

```python
#Dropping 4 unnecessary rows 
df = df.iloc[4:]
```

```python
df.head()
```

```python
#Creating a column Year by index
df['Year'] = df.index
df.head()
```

```python
# Reseting index
df.reset_index(inplace = True, drop = True)
```

```python
df.head()
```

```python
#getting the list about all countries
countries = df.columns.tolist()
```

```python
#transforming all countries columns to float type
df[countries] = df[countries].astype(float)
```

```python
countries
```

```python
#detecting data types
df.dtypes
```

```python
#transforming Year column to int type
df['Year'] = df['Year'].astype(int)
```

```python
df.head()
```

```python
#Creating a graph about the percentage of poverty in Europe Countries

plt.plot(df['Year'], df['United Kingdom'], label='United Kingdom')
plt.plot(df['Year'], df['France'], label='France')
plt.plot(df['Year'], df['Germany'], label='Germany')
plt.plot(df['Year'], df['Italy'], label='Italy')
plt.plot(df['Year'], df['Spain'], label='Spain')
plt.xlabel('Year')
plt.ylabel('Poverty % of population')
plt.title('Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population) x Europe Countries')
plt.grid(True)
plt.legend()
```

```python
#Creating a graph about the percentage of poverty in South American Countries

plt.plot(df['Year'], df['Brazil'], label='Brazil')
plt.plot(df['Year'], df['Argentina'], label='Argentina')
plt.plot(df['Year'], df['Colombia'], label='Colombia')
plt.plot(df['Year'], df['Venezuela, RB'], label='Venezuela, RB')
plt.plot(df['Year'], df['Paraguay'], label='Paraguay')
plt.xlabel('Year')
plt.ylabel('Poverty % of population')
plt.title('Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population) x South American Countries')
plt.grid(True)
plt.legend()
```

```python
#Creating a graph about the percentage of poverty in American Countries

plt.plot(df['Year'], df['United States'], label='United States')
plt.plot(df['Year'], df['Canada'], label='Canada')
plt.plot(df['Year'], df['Mexico'], label='Mexico')
plt.plot(df['Year'], df['Brazil'], label='Brazil')
plt.plot(df['Year'], df['Argentina'], label='Argentina')
plt.xlabel('Year')
plt.ylabel('Poverty % of population')
plt.title('Poverty headcount ratio at $1.90 a day (2011 PPP) (% of population) x American Countries')
plt.grid(True)
plt.legend()
```
